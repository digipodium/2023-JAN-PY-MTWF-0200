flipkart

```python
url = 'https://www.flipkart.com/search?q=bags'
sc = Scraper(url)
target = Tag('div',cls='_1YokD2 _3Mn1Gg')
items = Tag('div',cls='_1xHGtK _373qXS')
# to be scraped
brand = Tag('div', cls='_2WkVRV')
title = Tag('a', cls='IRpwTa')
sprice = Tag('div', cls='_30jeq3')
# execute
output = sc.get_all(target, items, brand=brand, title=title, sprice=sprice)
```
