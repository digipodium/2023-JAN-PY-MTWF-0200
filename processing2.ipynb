```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('laptops.csv')
df
```

```python
def get_brand_name(name):
    brand = name.split()[0]
    brand = brand.lower()
    return brand
```

```python
get_brand_name('RedmiBook Pro Core i5 11th Gen - (8 GB/512 GB')
```

```python
df['name'].apply(get_brand_name)
```

```python
df['brand'] = df['name'].apply(get_brand_name)
df
```

```python
def clean_price_column(price):
    if isinstance(price, str):
        price = price.replace('₹', '').replace(',', '')
        price = int(price)
    return price
```

```python
clean_price_column('₹2,41,990')
```

```python
df['price'].apply(clean_price_column)
```

```python
df['clean_price'] = df['price'].apply(clean_price_column)
df
```

```python
def get_num_of_rating(rr):
    if isinstance(rr, str):
        nor = rr.split()[0]
        nor = nor.replace(',', '')
        nor = int(nor)
        return nor
    else:
        np.nan

def get_num_of_reviews(rr):
    if isinstance(rr, str):
        x = rr.split('&')[-1]
        x = x.split()[0]
        x = x.replace(',', '')
        x = int(x)
        return x
    else:
        np.nan
```

```python
df['num_of_rating'] = df['rr'].apply(get_num_of_rating)
df['num_of_reviews'] = df['rr'].apply(get_num_of_reviews)
df
```
