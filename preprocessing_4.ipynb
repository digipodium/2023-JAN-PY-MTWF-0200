```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None) # when there are too many columns, show all of them
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/digipodium/Datasets/main/regression/automobile.csv')
df
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.replace('?', np.nan, inplace=True)
df.isnull().sum()
```

```python
num_cols_with_missing_data = ['normalized-losses', 'bore', 'stroke',
                              'horsepower', 'peak-rpm', 'price']
cat_cols_with_missing_data = ['num-of-doors']
df[num_cols_with_missing_data]
```

```python
for col in num_cols_with_missing_data:
    df[col] = df[col].astype(float)
    df[col].fillna(df[col].mean(), inplace=True)
df
```

```python
for col in cat_cols_with_missing_data:
    df[col].fillna(df['num-of-doors'].value_counts().idxmax(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns.tolist()
```

```python
df.rename(columns={
    'normalized-losses': 'nl',
    'num-of-doors': 'doors',
    'engine-type':'engine',
    },inplace=True)
df
```
